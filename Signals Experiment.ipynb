```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Signal
a = 10 # Amplitude
alpha = 0.65
n = np.array([x for x in np.arange(0, 15, 1)])
y = a * alpha ** n

# Plot figure
fig, ax = plt.subplots(figsize = (13, 5))
ax.set_xlabel('n'), ax.set_xlim(-1, 15)
ax.set_ylabel('Amplitude')
ax.set_title('Signal Plot: $Aα^{n}$ when  A is +ve and α is between 0 and 1')
ax.stem(n, y)
```

```python
# Signal
a = 10 # Amplitude
alpha2 = -0.65 # Negative exponential
n = np.array([x for x in np.arange(0, 15, 1)])
y = a * alpha2 ** n

# Plot figure
fig2, ax2 = plt.subplots(figsize = (13, 5))
ax2.set_xlabel('n'), ax.set_xlim(-1, 15)
ax2.set_ylabel('Amplitude')
ax2.set_title('Signal Plot: $Aα^{n}$ when  A is +ve and α is between -1 and 0')
ax2.stem(n, y)
```
